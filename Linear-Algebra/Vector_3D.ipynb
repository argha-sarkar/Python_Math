```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.gca(projection='3d')
vec = np.array([[7,-5],[7,2],[7,2]])
origin = np.zeros(vec.shape)
ax.quiver(*origin, *vec)

plt.grid()
ax.set_zlim3d(-10, 10)
ax.set_ylim3d(-10, 10)
ax.set_xlim3d(-10, 10)  
plt.show()
```
